1. Create a standalone function

```python
def get_popularity(x):
  if x < 220:
    return 'relatively unknown'
  elif x < 600:
    return 'kind of popular'
  elif x < 2000:
    return 'popular'
  else:
    return 'super-popular'
```

2. Apply the function from the step above to the players page_views column. Vectorized approach

```python
players.page_views.apply(get_popularity)
```

3. Add the output from the step above as a new column to the players dataframe. Name the column popularity.

```python
players['popularity'] = players.page_views.apply(get_popularity)
players.popularity
```

4. How many "super_popular" players are there?

```python
players[players.popularity == 'super-popular'].name.size
```

# The SettingWithCopy Warning
* SettingWithCopyWarnings
* pd.options.mode.chained_assignment = 'warn' --> default
* pd.options.mode.chained_assignment = 'None' --> to switch warning off

```python
players.head()
# alter Peter Chest's page views to 2001
```

```python
# Using [] indexing
players['page_views'][2] = 2001
# we get a warning as:
# <ipython-input-232-1a1c499a9ce6>:2: SettingWithCopyWarning: 
# A value is trying to be set on a copy of a slice from a DataFrame
```

```python
players.head()
```

```python
players.drop_duplicates()['page_views'][2] = 3000
# here we are doing chain assignment over a function that returns a copy of dataframe
```

```python
players.head()
# doe snot change the value
```

```python
# We can also switch off this warning setting the chained_assignment mode
# pd.options.mode.chained_assignment = 'warn'
# pd.options.mode.chained_assignment = 'None' -- we can also switch it off by setting to None
```
