# The sample() Method
* Always returns one random record/observation each time we call it
* .sample(random_state=n) --> Returns same record each time we query it
* .sample(n=record_count) --> Returns record_count random records
* .sample(frac=fraction_value) --> Returns fraction of teh total records
* n and frac parameters can't be used together in single sample() call

```python
nutrition.sample()
```

```python
nutrition.sample(random_state=12)
```

```python
nutrition.sample(n=3)
# .sample(n=1) ~ .sample()
```

```python
nutrition.sample(frac=0.01)
```
